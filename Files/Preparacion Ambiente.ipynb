```python

%sql
create widget text storageName default "adlsproyectofinalsd";
```

# Catalog

```python

%sql
CREATE CATALOG IF NOT EXISTS proyectofinal
MANAGED LOCATION 'abfss://data@${storageName}.dfs.core.windows.net/'
```

# Schemas

```python
%sql
CREATE SCHEMA IF NOT EXISTS proyectofinal.bronze;
CREATE SCHEMA IF NOT EXISTS proyectofinal.silver;
CREATE SCHEMA IF NOT EXISTS proyectofinal.golden;
```
