#Imports

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

#Widgets

```python
dbutils.widgets.removeAll()
```

```python
dbutils.widgets.text("storageName", "adlsproyectofinalsd")
dbutils.widgets.text("container", "raw")
dbutils.widgets.text("catalog", "proyectofinal")
dbutils.widgets.text("schema", "bronze")
```

# Variables

```python
storageName =  dbutils.widgets.get("storageName")
container = dbutils.widgets.get("container")
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
```

#Path

```python
path_artist = f"abfss://{container}@{storageName}.dfs.core.windows.net/artists.csv"
```

# Structures

```python
artist_schema = StructType(fields=[StructField("id", IntegerType(), False),
                                     StructField("name", StringType(), True),
                                     StructField("followers", StringType(), True),
                                     StructField("popularity", StringType(), True),
                                     StructField("genres", StringType(), True),
                                     StructField("main_genre", StringType(), True)
])
```

```python

df_artists = spark.read.option('header', True)\
                        .schema(artist_schema)\
                        .csv(path_artist)
```

```python
display(df_artists)
```

```python
df_artists.write.mode("overwrite").saveAsTable(f"{catalog}.{schema}.artists")
```
